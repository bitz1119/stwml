```python
Y = df2.values[:,1:]
```

```python
import numpy as np
```

```python
def KNN(X,Y,point,K=5):
    lis = []
    for i in range(X.shape[0]):
        dis = np.sqrt(np.sum((X[i]-point)**2))
        lis.append((dis,Y[i]))
        
    lis.sort()
    ans = lis[:K]
    count0 = 0
    count1 = 0
    
    for i in range(K):
        if(lis[i][1]==0):
            count0 += 1
        else:
            count1 += 1
            
    if(count1>count0):
        return 1
    else:
        return 0

    
        
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c = Y.flatten())
point  = np.array([1,3])
plt.scatter(point[0],point[1],c = "red")

KNN(X,Y,point,K=100)
```
