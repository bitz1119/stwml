```python
def k_mean_clustering():
    x_ke_pass_wale = []
    y_ke_pass_wale = []
    #must be generated according to the range of input values
    point = np.random.random((2,2))*6  
    for j in range(10):
        for i in range(X.shape[0]):
            x_se_dis = np.sqrt(np.sum((point[0]-X[i])**2))
            y_se_dis = np.sqrt(np.sum((point[1]-X[i])**2))
            if(x_se_dis < y_se_dis):
                x_ke_pass_wale.append(X[i])
            else:
                y_ke_pass_wale.append(X[i])

        x_ke_pass_wale =  np.array(x_ke_pass_wale)
        y_ke_pass_wale = np.array(y_ke_pass_wale)    
        point[0] = x_ke_pass_wale.mean(axis = 0)
        point[1] = y_ke_pass_wale.mean(axis =0)
        x_ke_pass_wale = []
        y_ke_pass_wale = []
        plt.scatter(X[:,0],X[:,1])
        plt.scatter(point[:,0],point[:,1])
        plt.show()
    return point
        
    
```
