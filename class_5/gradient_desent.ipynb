```python
import numpy as np
X = np.array([1,2,3,4,5,6])
Y =np.array([2,3,4,5,6,7])
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y)
plt.show()
```

```python
def gradient_decent(X,Y):
    m = 0
    c = 0
    learning_rate = 0.001
    for i in range(1000):
        cost = Y-(m*X+c)
        m = m + learning_rate*np.sum((cost*X))
        c = c + learning_rate*np.sum((cost))
        
    return m,c
        
```

```python
m,c = gradient_decent(X,Y)
```

```python
plt.scatter(X,Y)
plt.plot(X,m*X+c)
plt.show()
```
